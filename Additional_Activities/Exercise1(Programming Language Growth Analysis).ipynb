```python
import pandas as pd
import numpy as np

# Configuration
STUDENT_ID = 20251234  # Replace with your ID
df = pd.read_csv('most-popular-programming-languages-2004-2024.csv')
df['Month'] = pd.to_datetime(df['Month'])

# Language Assignment Logic [cite: 480, 482]
languages = sorted([col for col in df.columns if col != 'Month'])
language_index = (STUDENT_ID % 1000) % len(languages)
assigned_lang = languages[language_index]
print(f"Assigned Language: {assigned_lang}")

# Extract data [cite: 483]
data = df[['Month', assigned_lang]].copy()
data.rename(columns={assigned_lang: 'Popularity'}, inplace=True)

# 1. Growth Rate and Rolling Stats [cite: 484, 485]
data['Growth_Rate'] = data['Popularity'].pct_change() * 100
data['Moving_Avg'] = data['Popularity'].rolling(window=12).mean()
data['Moving_STD'] = data['Popularity'].rolling(window=12).std()

# 2. Classify Phases using NumPy logic [cite: 486]
threshold = 5.0
conditions = [
    (data['Growth_Rate'] > threshold),
    (data['Growth_Rate'] < -threshold),
    (data['Growth_Rate'].between(-threshold, threshold))
]
choices = ['Growth', 'Decline', 'Stable']
data['Phase'] = np.select(conditions, choices, default='Stable')

# Output Tables [cite: 490, 491, 492]
print("\n--- DataFrame Output (First 5 Rows) ---")
print(data.head(15))

print("\n--- Statistical Summary ---")
print(data['Popularity'].describe())

print("\n--- Phase Counts ---")
print(data['Phase'].value_counts())

# Key Performance Metrics [cite: 492]
initial_pop = data['Popularity'].iloc[0]
final_pop = data['Popularity'].iloc[-1]
overall_growth = ((final_pop - initial_pop) / initial_pop) * 100
print(f"\nOverall Growth: {overall_growth:.2f}%")
```

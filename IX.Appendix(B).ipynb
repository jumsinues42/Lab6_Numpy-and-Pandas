# 1. Using try-except Blocks for Safe NumPy Operations 

```python
import numpy as np 
try: 
 a = np.array([1, 2, 3]) 
 b = np.array([[1], [2], [3]]) 
 result = a + b 
 print(result) 
except ValueError as e: 
 print(f"Broadcasting error: {e}")
```

# 2. Creating and Using Identity Matrices in Computation 


```python
identity = np.eye(3) 
matrix = np.array([[2, 3, 4], [1, 0, 6], [7, 8, 9]]) 
print("Matrix + Identity:\n", matrix + identity) 
```

# 3. Detecting NaNs and Invalid Data in Arrays (NumPy)

```python
data = np.array([1, 2, np.nan, 4, np.inf]) 
cleaned = data[np.isfinite(data)] 
print("Cleaned Data:", cleaned) 
```

# 4. Handling Missing Data in Pandas

```python
import pandas as pd 
df = pd.DataFrame({ 
 "Name": ["Alice", "Bob", "Charlie", None], 
 "Score": [85, None, 92, 88] 
}) 
df_cleaned = df.dropna() 
print("Cleaned DataFrame:\n", df_cleaned) 
```

# 5. Applying NumPy Functions to Pandas Columns 

```python
df = pd.DataFrame({ 
 "X": [1, 2, 3], 
 "Y": [4, 5, 6] 
}) 
df["Log_X"] = np.log(df["X"]) 
df["Sqrt_Y"] = np.sqrt(df["Y"]) 
print("Transformed DataFrame:\n", df) 
```
